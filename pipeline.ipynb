```python
import base64
from io import BytesIO
from PIL import Image

def decode_base64_uri_to_image(base64_uri, output_path="decoded_image.png"):
    """解码 Base64 URI 并保存为 PNG 图片"""
    # 确保 Base64 URI 以 'data:image/png;base64,' 开头
    if base64_uri.startswith("data:image"):
        base64_data = base64_uri.split(",", 1)[1]  # 取出真正的 Base64 数据
    else:
        base64_data = base64_uri  # 如果没有前缀，直接使用

    # 解码 Base64
    image_data = base64.b64decode(base64_data)
    
    # 用 PIL 读取图片
    image = Image.open(BytesIO(image_data))

    # 保存图片到本地
    image.save(output_path, "PNG")
    print(f"图片已成功解码并保存为 {output_path}")

# 示例 Base64 URI（你的真实数据）
base64_uri = test_uri
# 运行解码
decode_base64_uri_to_image(base64_uri, "test_output.png")
```
